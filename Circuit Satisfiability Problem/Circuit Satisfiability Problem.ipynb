```python
import matplotlib.pyplot as plt
```

```python
Tijden = [409, 365, 326, 324, 323]
Processen = ['1','2','4','8', '16']
```

```python
def grafiek():
    plt.plot(Processen, Tijden, label = "Tijd")
    plt.xlabel('Processen')
    plt.ylabel('Duur in seconden')
    plt.title('tijdsduur met verschillende hoeveelheden processen')
    plt.legend()
```

```python
grafiek()
```

De bovenstaande grafiek geeft de tijd weer die het duurt om 1,2,4,8 en 16 processen te runnen van circuitSatifiability.c
Uit de grafiek is te lezen dat hoe meer processen er gebeuren hoe minder tijd er nodig is. Alhoewel, dat is aan het begin het geval, naarmate er meer processen bijkomen neemt de tijd steeds minder snel af. Het neemt zelfs zoveel af dat het eigenlijk hetzelfde blijft na een tijdje. Dat is ook te zien als we het procentuele verschil gaan berekenen, dat doen we met behulp van de volgende formule: (nieuw-oud) / oud. Het verschil gaat van 10.7% naar 10.6% naar opeens 0.6% en 0.3%.
